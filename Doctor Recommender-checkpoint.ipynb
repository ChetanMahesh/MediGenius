```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Load the datasets
train_data = pd.read_csv("Training.csv")
test_data = pd.read_csv("Testing.csv")
doctors_data = pd.read_csv("doctors_dataset.csv")

# Explore the dataset to identify features and target columns
print("Training Data Head:\n", train_data.head())
print("Testing Data Head:\n", test_data.head())
print("Doctors Dataset Head:\n", doctors_data.head())

# Assuming the last column is the target label
X_train = train_data.iloc[:, :-1]
y_train = train_data.iloc[:, -1]

X_test = test_data.iloc[:, :-1]
y_test = test_data.iloc[:, -1]

# Train a simple classifier (Random Forest)
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")

# Plot confusion matrix
cm = confusion_matrix(y_test, y_pred)
cmd = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)

# Plot the confusion matrix with increased size and adjusted label rotation
fig, ax = plt.subplots(figsize=(18, 12))
cmd.plot(cmap=plt.cm.Blues, ax=ax, colorbar=True)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels
plt.yticks(rotation=45)  # Rotate y-axis labels
plt.title("Confusion Matrix")
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()

# Plot model accuracy and loss (example graph structure)
history = {'accuracy': [0.82, 0.85, 0.88, 0.90, 0.92], 'val_accuracy': [0.80, 0.83, 0.87, 0.89, 0.91],
           'loss': [0.4, 0.35, 0.30, 0.25, 0.2], 'val_loss': [0.42, 0.38, 0.32, 0.28, 0.23]}

acc = history['accuracy']
val_acc = history['val_accuracy']
loss = history['loss']
val_loss = history['val_loss']
epochs_range = range(1, len(acc) + 1)

# Create a figure with subplots
plt.figure(figsize=(15, 5))

# Plot Accuracy
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Train Set', color='blue')
plt.plot(epochs_range, val_acc, label='Val Set', color='orange')
plt.legend(loc="best")
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Model Accuracy')

# Plot Loss
plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Train Set', color='blue')
plt.plot(epochs_range, val_loss, label='Val Set', color='orange')
plt.legend(loc="best")
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Model Loss')

# Display the plots
plt.tight_layout()
plt.show()
```
